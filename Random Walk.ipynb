```python
"""
Created on Fri Apr 17 15:29:05 2020

@author: ADIV
"""
import numpy
import matplotlib.pyplot as plt
```

```python
iteration = 100
particle = 10
x = numpy.zeros(iteration)
y = numpy.zeros(iteration)
arrayX = numpy.zeros((iteration,particle))
arrayY = numpy.zeros((iteration,particle))

for i in range(iteration):
    for j in range(particle):
        random = numpy.random.uniform(low = 0,high = 1)
        if random <= 0.25:
            x[j] += 1
        elif 0.25 < random <= 0.5:
            x[j] -= 1
        elif 0.5 < random <= 0.75:
            y[j] += 1
        else:
            y[j] -= 1
        arrayX[i][j] = x[j]
        arrayY[i][j] = y[j]

color = ['red','green','blue','cyan','black','yellow','grey','magenta','orange','purple']
plt.xlabel('X')
plt.ylabel('Y')
for i in range(particle):
    plt.plot(arrayX[i],arrayY[i],c = color[i%9])
    titikAwalX = [arrayX[i,0],arrayX[i,-1]]
    titikAwalY = [arrayY[i,0],arrayY[i,-1]]
    plt.scatter(titikAwalX,titikAwalY,c = color[i%9])
    plt.plot(titikAwalX,titikAwalY,c = color[i%9],linestyle = '-')

plt.show()

distance = []

for i in range (particle):
    X = x[i] ** 2
    Y = y[i] ** 2
    distance.append(X+Y)

print('Jarak Maksimum = ',max(distance))
print('Jarak Minimum  = ',min(distance))
print('Jarak Rata - Rata = ',numpy.average(distance))
```
